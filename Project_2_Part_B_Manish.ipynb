```python
# Performing image pre-processing before inputting it to the convolutional neural network


    
```

```python
from PIL import Image

max_age = dp['Age'].max()
def generate_image_data(data_index):
    
    age_l, sex_l, race_l, image_l = [],[],[],[]
    while True:
        for i in data_index:
            instance = dp.iloc[i]
            age = instance['Age']
            race = instance['Race']
            sex = instance['Sex']
            img_path = instance['Path']
            
            ima = Image.open(img_path)
            ima = ima.resize((im_width, im_height))
            ima = np.array(ima) / 255.0
            
            age_l.append(age / max_age)
            race_l.append(race)
            sex_l.append(sex)
            image_l.append(ima)
            
```

# Building our CNN model

```python
from keras.models import Model
#from keras.layers.normalization import BatchNormalization
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers.core import Activation
from keras.layers.core import Dropout
from keras.layers.core import Lambda
from keras.layers.core import Dense
from keras.layers import Flatten
from keras.layers import Input
import tensorflow as tf
```

```python
class mainModel():
    
    def hidden_layers(self, inputs):
        x = Conv2D(14, (3, 3), padding="same")(inputs)
        x = Activation("relu")(x)
        #x = BatchNormalization(axis=-1)(x)
        x = MaxPooling2D(pool_size=(3, 3))(x)
        x = Dropout(0.2)(x)

        x = Conv2D(30, (3, 3), padding="same")(x)
        x = Activation("relu")(x)
        #x = BatchNormalization(axis=-1)(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        x = Dropout(0.2)(x)

        x = Conv2D(30, (3, 3), padding="same")(x)
        x = Activation("relu")(x)
        #x = BatchNormalization(axis=-1)(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        x = Dropout(0.2)(x)

        return x
    
    def age_model(self, inputs):
        
        x = self.hidden_layers(inputs)

        x = Flatten()(x)
        x = Dense(120)(x)
        x = Activation("relu")(x)
        #x = BatchNormalization()(x)
        x = Dropout(0.5)(x)
        x = Dense(1)(x)
        x = Activation("linear", name="age_result")(x)

        return x
    
    def sex_model(self, inputs, no_of_sex=2):
        
        x = Lambda(lambda c: tf.image.rgb_to_grayscale(c))(inputs)

        x = self.hidden_layers(inputs)

        x = Flatten()(x)
        x = Dense(130)(x)
        x = Activation("relu")(x)
        #x = BatchNormalization()(x)
        x = Dropout(0.5)(x)
        x = Dense(no_of_sex)(x)
        x = Activation("sigmoid", name="sex_result")(x)
        
        return x
    
    def race_model(self, inputs, no_of_races = 5):
        x = self.hidden_layers(inputs)

        x = Flatten()(x)
        x = Dense(130)(x)
        x = Activation("relu")(x)
        #x = BatchNormalization()(x)
        x = Dropout(0.5)(x)
        x = Dense(no_of_races)(x)
        x = Activation("softmax", name="race_RESULT")(x)

        return x
    
    def multi_model(self, width, height, no_of_races):
        
        input_shape = (height, width, 3)

        inputs = Input(shape=input_shape)

        age_r_model = self.age_model(inputs)
        race_r_model = self.race_model(inputs, no_of_races)
        sex_r_model = self.sex_model(inputs)

        model = Model(inputs=inputs,
                     outputs = [age_r_model, race_r_model, sex_r_model],
                     name="Project-2-Task-1")

        return model
```

```python
#Calling the multi model
model = mainModel().multi_model(im_width, im_height, no_of_races = 5)
```

# Training our model

```python
from tensorflow.keras.optimizers import Adam

init_lr = 1e-4
epochs = 100

opt = Adam(lr=init_lr, decay=init_lr / epochs)

model.compile(optimizer=opt, 
              loss={
                  'age_result': 'mse', 
                  'race_result': 'categorical_crossentropy', 
                  'sex_result': 'binary_crossentropy'},
              loss_weights={
                  'age_result': 4., 
                  'race_result': 1.5, 
                  'sex_result': 0.1},
              metrics={
                  'age_result': 'mae', 
                  'race_result': 'accuracy',
                  'sex_result': 'accuracy'})
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

train_gen = generate_image_data(train_index)
valid_gen = generate_image_data(valid_index)

callbacks = [
    ModelCheckpoint("./model_checkpoint", monitor='val_loss')
]

history = model.fit_generator(train_gen,
                    steps_per_epoch=len(train_index)//100,
                    epochs=epochs,
                    callbacks=callbacks,
                    validation_data=valid_gen,
                    validation_steps=len(valid_index)//100)
```
